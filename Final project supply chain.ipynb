```python
import pandas as pd
# Import necessary libraries
from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt


# Display the first few rows of the dataset
print(data.head())

# Filter rows where 'portname' is 'Aabenraa'
data = data[data['portname'] == 'Porto De Suape']

# Get the number of rows in the filtered dataset
num_rows = data.shape[0]

print(f"The dataset has {num_rows} rows.")

# Display the filtered dataset
print(data.head())


# Calculate total daily volume (import + export)
data['total_volume'] = data['import_cargo'] + data['export_cargo']

# Display the updated dataset
print(data[['date', 'import_cargo', 'export_cargo', 'total_volume']].head())


# Get the number of rows in the filtered dataset
num_rows = data.shape[0]

print(f"The dataset has {num_rows} rows.")


import matplotlib.pyplot as plt
import pandas as pd

# Ensure 'date' is in datetime format for proper plotting
data['date'] = pd.to_datetime(data['date'])

# Plot total volume over time
plt.figure(figsize=(12, 6))
plt.scatter(data['date'], data['total_volume'], label='Total Volume', color='blue', linewidth=1.5)
plt.title('Total Volume Over Time', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Total Volume (Metric Tons)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()


import matplotlib.pyplot as plt

# Filter the data for a specific year (e.g., 2023)
filtered_data = data[data['date'].dt.year == 2023]

# Plot total volume for the filtered year
plt.figure(figsize=(12, 6))
plt.scatter(filtered_data['date'], filtered_data['total_volume'], label='Total Volume', color='blue', linewidth=1.5)
plt.title('Total Volume Over 2023', fontsize=16)
plt.xlabel('Date', fontsize=14)
plt.ylabel('Total Volume (Metric Tons)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()





# Filter data for two years (e.g., 2022 and 2023)
multi_year_data = data[(data['date'].dt.year >= 2021) & (data['date'].dt.year <= 2023)]

# Ensure 'date' is the index and sorted
multi_year_data = multi_year_data.set_index('date').sort_index()

# Perform seasonal decomposition with a yearly period
decomposition = seasonal_decompose(multi_year_data['total_volume'], model='additive', period=365)

# Plot the components
decomposition.plot()
plt.rcParams.update({'figure.figsize': (12, 8)})  # Adjust plot size
plt.show()


# Filter data for years 2021 to 2023
filtered_data = data[(data['date'].dt.year >= 2021) & (data['date'].dt.year <= 2023)]

# Aggregate 'total_volume' to weekly level while preserving other columns
weekly_data = filtered_data.resample('W', on='date').agg({
    'total_volume': 'sum',  # Aggregate total volume by summing
    'import_cargo': 'sum',  # Aggregate imports
    'export_cargo': 'sum',  # Aggregate exports
    # Add any other columns and aggregation methods as needed
})

# Ensure the index is properly named and reset (optional)
weekly_data = weekly_data.reset_index()

# Display the first few rows of the aggregated data
print(weekly_data.head())

# Perform seasonal decomposition on weekly total volume
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(weekly_data['total_volume'], model='additive', period=52)

# Plot the decomposition components
decomposition.plot()
plt.rcParams.update({'figure.figsize': (12, 8)})
plt.show()

# Plot the level with week annotations
plt.figure(figsize=(12, 6))
plt.plot(weekly_data['date'], weekly_data['total_volume'], label='Total Volume', color='blue')
plt.title('Weekly Total Volume (2022-2024)')
plt.xlabel('Date')
plt.ylabel('Total Volume')
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()
plt.tight_layout()
plt.show()



from statsmodels.tsa.holtwinters import ExponentialSmoothing

# Fit Holt-Winters model with weekly seasonality
model = ExponentialSmoothing(
    weekly_data['total_volume'], 
    seasonal='additive', 
    seasonal_periods=52  # Weekly seasonality
)
hw_model = model.fit()

# Forecast the next 52 weeks
forecast_hw = hw_model.forecast(steps=52)

# Print the forecasted weekly demands for 2025
print("Forecasted Weekly Demands for 2024 (Holt-Winters):")
print(forecast_hw)

# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(weekly_data['total_volume'], label='Historical Data', color='blue')
plt.plot(range(len(weekly_data), len(weekly_data) + 52), forecast_hw, label='Holt-Winters Forecast', color='green')
plt.title('Holt-Winters Forecast for Weekly Demands (2024)')
plt.xlabel('Week')
plt.ylabel('Total Volume')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()


# Extract level, trend, and seasonal components from Holt-Winters model
level = hw_model.level
trend = hw_model.trend
seasonal = hw_model.season

# Define the seasonality period (weekly seasonality)
seasonal_period = 52

# Compute the forecast column
forecast_column = []
for i in range(len(weekly_data)):
    # Calculate forecast for each time step
    forecast_value = level[i] + trend[i] + seasonal[i % seasonal_period]
    forecast_column.append(forecast_value)

# Add the forecast column to the dataset
weekly_data['forecast'] = forecast_column

# Display the updated dataset with the forecast column
print(weekly_data[['date', 'total_volume', 'forecast']].head())

# Filter data for 2024
forecast_2024 = data[(data['date'].dt.year == 2024)]

# Aggregate 'total_volume' to weekly totals
forecast_2024 = forecast_2024.resample('W', on='date').agg({
    'total_volume': 'sum'
}).reset_index()

# Check the first few rows to ensure it worked
print(forecast_2024.head())


# Extract the total_volume column for both forecasts
original_forecast = forecast_2024['total_volume'].values  # Real data for 2024
holt_winters_forecast = forecast_hw.values  # Holt-Winters forecast for 2024

# Extract the total_volume column for both forecasts
original_forecast = forecast_2024['total_volume'].values
holt_winters_forecast = weekly_data['total_volume'].values  # Use Holt-Winters generated forecast


# Extract the total_volume column for both forecasts
original_forecast = forecast_2024['total_volume'].values  # Real data for 2024

# Align Holt-Winters forecast length with the real data (first 45 weeks only)
holt_winters_forecast = forecast_hw[:len(original_forecast)]  # Slice to match length

# Ensure shapes match
print(f"Original Forecast Length: {len(original_forecast)}")
print(f"Holt-Winters Forecast Length: {len(holt_winters_forecast)}")


# Calculate MAE and MPE
mae = np.mean(np.abs(original_forecast - holt_winters_forecast))
mpe = np.mean((original_forecast - holt_winters_forecast) / original_forecast) * 100

# Print results
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Percentage Error (MPE): {mpe:.2f}%")

import matplotlib.pyplot as plt
import numpy as np

# Plot both forecasts
plt.figure(figsize=(12, 6))
plt.plot(original_forecast, label='Original Forecast', color='blue', marker='o')
plt.plot(holt_winters_forecast, label='Holt-Winters Forecast', color='orange', linestyle='--', marker='x')
plt.title('Comparison of Total Volume Forecasts')
plt.xlabel('Week Index')
plt.ylabel('Total Volume')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Calculate MAE and MPE
mae = np.mean(np.abs(original_forecast - holt_winters_forecast))
mpe = np.mean((original_forecast - holt_winters_forecast) / original_forecast) * 100

# Print results
print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Percentage Error (MPE): {mpe:.2f}%")
```
